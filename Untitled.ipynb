To update the prior, which is a beta distribution, here is the definition of the beta probability density function:

$$ \int (x, \alpha, \beta) = \frac { 1 } { B(\alpha, \beta) } x^{ \alpha - 1} (1 - x)^{ \beta - 1} $$

From this [post](https://archive.ph/Q6qO8#selection-1325.76-1353.113)
>Think of  and  as imaginary coin flips before you actually flip the coin:  - 1 is the number of heads you get, and  - 1 is the number of tails. If you imagine that you have flipped 18 times, and got 9 heads and 9 tails ( =  = 10), you are more confident that the coin is fair, than someone who imagines only 4 trials that resulted in 2 heads and 2 tails ( =  = 3). The reason we need to subtract 1 is that, by setting  =  = 1, we reduce numerator to a constant 1. This gives us the uninformative prior  we assume zero imaginary coin flips.

In the case of 20 flips, and 10 heads. Let's start with the uninformed prior of $\alpha = 1$ and $\beta = 1$ (uniformed distribution, essentially). Updating the prior so that:

$$ \alpha^* = \alpha + heads = 11 $$
$$ \beta^* = \beta + tails = 11 $$

```python
alpha = 11
beta = 11

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(probs, st.beta.pdf(probs, a=alpha, b=beta))
ax.set_xlabel("p")
ax.set_ylabel("density")
ax.set_title(f" = {alpha},  = {beta}")
    
```

```python
alphas = [1, 1, 4, 100]
betas = [1, 1/2, 4, 100]
n_trials = 20
k_success = 10
k_failure = n_trials - k_success

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for ax, alpha, beta in zip(axes.ravel(), alphas, betas):
    ax.plot(probs, st.beta.pdf(probs, a=alpha, b=beta), label="prior", linestyle="dotted")
    ax.plot(probs, st.beta.pdf(probs, a=k_success + 1, b=k_failure + 1), label="likelihood", linestyle="--")
    ax.plot(probs, st.beta.pdf(probs, a=alpha + k_success, b=beta + k_failure), label="posterior")
    ax.set_xlabel("p")
    ax.set_ylabel("density")
    ax.set_title(f" = {alpha},  = {beta}")
axes.ravel()[0].legend();
```

**Bayes Factor** (BF) is the marginal likelihood  that is, the probability of data under a specified model $P(D|H)$  for one model divided by the marginal likelihood of another model:

$$ BF_{ 01 } = \frac { P(D|H_0) }{ P(D|H_1) } $$

It implies how probable one model is true over another for a certain belief.

In the case of the flipped coin, let yes our prior model is that of the uninformed, so $B(p = 0.5, 1, 1)$, and the posterior is that of the 10 heads and 10 tails case, so $B(p = 0.5, 11, 11)$:

$$ BF_{ 10 } = \frac { P(D|H_1) }{ P(D|H_0) } = \frac { B(p = 0.5, 11, 11) }{ B(p = 0.5, 1, 1) } = \frac { 3.70 }{ 1 } = 3.70 $$

This $BF_{10}$ tells at how likely we believe the coin is fair ($p = 0.5$, 50% heads or tails) between the prior held belief (prior model) and the belief after we see the coin flip outcomes (posterior model).

```python
st.beta.pdf(0.5, 11, 11), st.beta.pdf(0.5, 1, 1), st.beta.pdf(0.5, 11, 11) / st.beta.pdf(0.5, 1, 1)
```

```python
alphas = [1, 1, 4, 100]
betas = [1, 1/2, 4, 100]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
n_trials = 20
k_success = 10
k_failure = n_trials - k_success

for ax, alpha, beta in zip(axes.ravel(), alphas, betas):
    ax.plot(probs, st.beta.pdf(probs, a=alpha, b=beta), label="prior", linestyle="dotted")
    ax.plot(probs, st.beta.pdf(probs, a=k_success + 1, b=k_failure + 1), label="likelihood", linestyle="--")
    ax.plot(probs, st.beta.pdf(probs, a=alpha + k_success, b=beta + k_failure), label="posterior")
    ax.plot((0.5, 0.5), (st.beta.pdf(0.5, a=alpha + k_success, b=beta + k_failure), st.beta.pdf(0.5, a=alpha, b=beta)), linestyle="dotted", color="black")
    ax.set_xlabel("p")
    ax.set_ylabel("density")
    ax.set_title(f" = {alpha},  = {beta}, Bayes factor at p = 0.5: {st.beta.pdf(0.5, a=alpha + k_success, b=beta + k_failure) / st.beta.pdf(0.5, a=alpha, b=beta):.2f}")
axes.ravel()[0].legend();
```

The mean of the posterior distribution (for a Beta distribution) can be calculated by $\frac {\alpha}{\alpha + \beta}$.

```python
alphas = [1, 1, 4, 100]
betas = [1, 1/2, 4, 100]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
n_trials = 100
k_success = 90
k_failure = n_trials - k_success

for ax, alpha, beta in zip(axes.ravel(), alphas, betas):
    ax.plot(probs, st.beta.pdf(probs, a=alpha, b=beta), label="prior", linestyle="dotted")
    ax.plot(probs, st.beta.pdf(probs, a=k_success + 1, b=k_failure + 1), label="likelihood", linestyle="--")
    ax.plot(probs, st.beta.pdf(probs, a=alpha + k_success, b=beta + k_failure), label="posterior")
    ax.axvline((alpha + k_success) / (alpha + k_success + beta + k_failure), linestyle="dotted", color="black")
    ax.set_xlabel("p")
    ax.set_ylabel("density")
    ax.set_title(f" = {alpha},  = {beta}, Bayes factor at p = 0.5: {st.beta.pdf(0.5, a=alpha + k_success, b=beta + k_failure) / st.beta.pdf(0.5, a=alpha, b=beta):.2f}")
axes.ravel()[0].legend();
```
